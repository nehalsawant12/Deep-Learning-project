```python
X=np.array([1,2,3,4])
Y=np.array([5,6,7,8])
```

```python
#Create a function for grediant descent 
def gradient_descent(X,Y):
    n=len(X)
    m_curr=0 #suppose initialize slope means gradient means coefficient 
    #means m=0
    b_curr=0  #suppose initialize intercept means b=0
    iterations=100 #user defined variable iterations
    learning_rate=0.01 #assign (alpha or lambda or nitta)
    print("Slope : ",m_curr,"Intercept : ",b_curr," First Time")
    plt.scatter(X,Y,color="red")
    plt.xlabel("Value of X ")
    plt.ylabel("Value  Of Y" )
    for i in range(iterations) : #0 to 99
        Y_pred=m_curr*X+b_curr #Equation of predicted line Y_pred=mX+b
        plt.plot(X,Y_pred,color='green')
        #find mean squared error 
        mse=(1/n)*sum([val**2 for val in (Y-Y_pred)])
        #Minimize the error(mse : cost or loss function), 
        #so differentiate mse with respect to slope m and intercept b
        m_diff=-(2/n)*sum(X*(Y-Y_pred))
        b_diff=-(2/n)*sum(Y-Y_pred)
        #New value of slope and intercept (change the value of m_curr and b_curr)
        m_curr=m_curr-learning_rate*m_diff
        b_curr=b_curr-learning_rate*b_diff
        print("Iteration No. : ",i+1)
        print("Slope : ",m_curr,"Intercept : ",b_curr)
        print("MSE : ",mse)
    return Y_pred
```

```python
#call in main program
Y_pred=gradient_descent(X,Y)
df['Predicted_Output']=Y_pred
df
```
